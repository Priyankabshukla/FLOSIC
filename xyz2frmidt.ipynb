```python
up=0
dn=0
up_fod=[]
dn_fod=[]
infile=open(path,'r')
for i,line in enumerate(infile):
    if len(line.split())==4:
#         print('Hi',i,line.split(),line.split()[0])
        if str(line.split()[0])=='He':
            print('Hi',i,line.split(),line.split()[0])
            up_fod.append([str(line.split()[0]),float(line.split()[1])*1.88973,float(line.split()[2])*1.88973,float(line.split()[3])*1.88973])
            up+=1
#             print('Hi',i,line.split())
        elif str(line.split()[0])=='X':
            dn_fod.append([str(line.split()[0]),float(line.split()[1])*1.88973,float(line.split()[2])*1.88973,float(line.split()[3])*1.88973])
            dn+=1
```
